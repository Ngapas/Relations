```python
# Given data
A = {1, 2, 3, 4}
B = {'x', 'y', 'z'}
C = {'a', 'b', 'c'}
R = {(1, 'b'), (2, 'a'), (2, 'c')}
S = {('a', 'y'), ('b', 'x'), ('c', 'y'), ('c', 'z')}

# Function to find the composition relation R ∘ S
def composition(R, S):
    result = set()
    for a, b1 in R:
        for b2, c in S:
            if b1 == b2:
                result.add((a, c))
    return result

# Function to find R^-1
def find_inverse(R):
    return {(y, x) for x, y in R}

# Function to find S^-1
def find_inverse_S(S):
    return {(y, x) for x, y in S}

# Function to determine the domain and range of a relation
def find_domain_and_range(relation):
    domain = {x for x, _ in relation}
    range_ = {y for _, y in relation}
    return domain, range_

# Function to check if a relation is reflexive
def is_reflexive(relation, A):
    return all((a, a) in relation for a in A)

# Function to check if a relation is symmetric
def is_symmetric(relation):
    return all((b, a) in relation for a, b in relation)

# Perform the operations
R_composition_S = composition(R, S)
R_inverse = find_inverse(R)
S_inverse = find_inverse_S(S)
R_domain, R_range = find_domain_and_range(R)
S_domain, S_range = find_domain_and_range(S)
is_reflexive_R = is_reflexive(R, A)
is_symmetric_R = is_symmetric(R)
is_reflexive_S = is_reflexive(S, B)
is_symmetric_S = is_symmetric(S)

print("R ∘ S =", R_composition_S)
print("R^-1 =", R_inverse)
print("S^-1 =", S_inverse)
print("Domain of R =", R_domain)
print("Range of R =", R_range)
print("Is R reflexive?", is_reflexive_R)
print("Is R symmetric?", is_symmetric_R)
print("Domain of S =", S_domain)
print("Range of S =", S_range)
print("Is S reflexive?", is_reflexive_S)
print("Is S symmetric?", is_symmetric_S)
```
